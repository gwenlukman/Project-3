```python
import matplotlib.pyplot as plt
import requests
import pandas as pd
from configg import api_key
```

```python
url = 'http://api.weatherbit.io/v2.0/history/daily'
```

```python
response = requests.get(f"{url}?city=Oakland,CA&start_date=2023-01-01&end_date=2023-12-31&key={api_key}").json()
response
```

```python
# Convert the data to a pandas DataFrame
df = pd.DataFrame(response['data'])
# Write the DataFrame to a CSV file
df.to_csv('Oakland_weather_data.csv', index=False)
```

```python
northcal = pd.read_csv('../Data/Northern_California_Sample.csv')
northcal_df = pd.DataFrame(northcal)
northcal_df
```

```python
target_city = 'Oakland, CA'
oakland_df = northcal_df[northcal_df['name'] == target_city]
oakland_df
```

```python
oakland_df.dtypes
```

```python
oakland_weather = pd.read_csv('../Data/Oakland_weather_data.csv')
oakland_weather_df = pd.DataFrame(oakland_weather)
oakland_weather_df.rename(columns={'datetime': 'collection_date'}, inplace=True)
oakland_weather_df
```

```python
oakland_temp = oakland_weather_df[['collection_date', 'max_temp', 'min_temp']]
oakland_temp['average_temp'] = oakland_temp[['max_temp', 'min_temp']].mean(axis=1)
oakland_avetemp = oakland_temp[['collection_date', 'average_temp']]
oakland_avetemp
```

```python
merged_df = pd.merge(oakland_df, oakland_avetemp, on='collection_date', how='inner')
merged_df
```

```python
merged_df.to_csv('Oakland_finaldata.csv', index=False)
```
